```python
# 왼쪽 절반 ROI 좌표 계산 (D:\project\khuda_final\data\example2.MP4)
import cv2

video_path = r"D:\project\khuda_final\data\example2.MP4"
cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    raise FileNotFoundError(f"Cannot open video: {video_path}")

ret, frame = cap.read()
cap.release()
if not ret or frame is None:
    raise RuntimeError("Cannot read first frame from video.")

h, w = frame.shape[:2]
x, y, roi_w, roi_h = 0, 0, w // 2, h  # 왼쪽 절반
roi = {"x": x, "y": y, "width": roi_w, "height": roi_h}
corners = [(x, y), (x + roi_w, y), (x + roi_w, y + roi_h), (x, y + roi_h)]

print("ROI (left half):", roi)
print("Corners:", corners)

# 미리보기 이미지 저장 (선택)
preview = frame.copy()
cv2.rectangle(preview, (x, y), (x + roi_w - 1, y + roi_h - 1), (0, 255, 0), 2)
cv2.imwrite("left_half_preview.png", preview)
print("Saved preview to left_half_preview.png")
```
